## Importing library for use

```python
import pyttsx3
```

## Setting engine and preparing for use

```python
engine = pyttsx3.init()
voices = engine.getProperty('voices')
engine.setProperty('voice',voices[1].id) # voices[0] for male
```

### Reading text

```python
engine.say('Hello, Seth')
engine.runAndWait()
```
